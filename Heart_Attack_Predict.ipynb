```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import mean_squared_error, r2_score

# Drop the first unnamed column which is likely an ID and causes issues with scaling
df_processed = hap_data.drop(hap_data.columns[0], axis=1)

# Split 'Blood Pressure' into 'Systolic_Pressure' and 'Diastolic_Pressure'
df_processed[['Systolic_Pressure', 'Diastolic_Pressure']] = df_processed['Blood Pressure'].str.split('/', expand=True)

# Convert to numeric, coercing errors to NaN
df_processed['Systolic_Pressure'] = pd.to_numeric(df_processed['Systolic_Pressure'], errors='coerce')

y_class = df_processed['Heart Attack Risk']        # classification target
y_reg = df_processed['Systolic_Pressure']  # regression target: using Systolic Pressure for now

X = df_processed.drop(['Heart Attack Risk', 'Blood Pressure', 'Systolic_Pressure', 'Diastolic_Pressure'], axis=1)

# Identify categorical columns for one-hot encoding
categorical_cols = X.select_dtypes(include='object').columns
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Align X_encoded and y_reg after potential NaN removal
# Remove rows where y_reg is NaN (due to coercion) from both X_encoded and y_reg
valid_indices = y_reg.dropna().index
X_encoded = X_encoded.loc[valid_indices]
y_reg = y_reg.loc[valid_indices]

X_train, X_test, y_train_c, y_test_c = train_test_split(
    X_encoded, y_class.loc[valid_indices], test_size=0.2, random_state=42
)

_, _, y_train_r, y_test_r = train_test_split(
    X_encoded, y_reg, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(X_train, y_train_c)

y_pred = model_lr.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test_c, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(X_train, y_train_c)

y_pred = model_dt.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test_c, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train_c)

y_pred = model_rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test_c, y_pred))
```

```python
from sklearn.svm import SVC

model_svm = SVC(kernel='rbf')
model_svm.fit(X_train, y_train_c)

y_pred = model_svm.predict(X_test)
print("SVM Accuracy:", accuracy_score(y_test_c, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train_c)

y_pred = model_knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test_c, y_pred))
```
